# Alphabet Rangoli

# Ans:

```python
def print_rangoli(size):
    li=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
    li1=li[:size][::-1]
    for i in range(1,len(li1)+1):
        o=li1[:i]+li1[:i][::-1]
        o.pop(len(o)//2)
        print('-'.join(o).center((((size*2)-1)*2)-1,'-'))
    for i in range(1,len(li1))[::-1]:
        o=li1[:i]+li1[:i][::-1]
        o.pop(len(o)//2)
        print('-'.join(o).center((((size*2)-1)*2)-1,'-'))

```

```python
print_rangoli(10)
```
